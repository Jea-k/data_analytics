#### How many restaurants are active on deliverando or our competitors in the given months (>0 orders per month)??

```python
# filtering rows from the df_dvr where kpi contains the word orders
dvr_orders = df_dvr[ (df_dvr['kpi'].str.contains('Orders'))]
```

```python
# dvr_orders = df_dvr[ (df_dvr['kpi'].str.contains('Orders'))]
```

```python
dvr_orders
```

```python
# Filtering for orders greater than 0 for each month for deliverando
active_res_dvr = dvr_orders[ (dvr_orders['month_1'] > 0) | (dvr_orders['month_2'] > 0) ]
```

```python
active_res_dvr
```

#### How much have the respective platforms grown?

```python
sum_1 = active_res_dvr['month_1'].sum()
sum_2 = active_res_dvr['month_2'].sum()
```

```python
dvr_growth = ((sum_2 - sum_1) / sum_1 ) *100
```

```python
print("% Growth by Deliverando:", dvr_growth)
```

```python
sum_cmp_1 = (active_res_cmp['month']==1).sum()
sum_cmp_2 = (active_res_cmp['month']==2).sum()
```

```python
cmp_growth = ((sum_cmp_2 - sum_cmp_1)/sum_cmp_1)*100
```

```python
print("% Growth by Competitors:", cmp_growth)
```

#### What conclusions regarding other KPIs can you draw from the performance of the top restaurants on deliverando?

```python
# Group by 'name' and 'zip' and calculate the sum of 'month_2' for each KPI
kpi_analysis = df_dvr.groupby(['name', 'zip', 'kpi'])['month_2'].sum().unstack(fill_value=0) # "unstacks" the grouped data.creates a new df with the 'kpi' values as columns

# Reset the index to get the 'name' and 'zip' back as columns
kpi_analysis.reset_index(inplace=True)

# Fill any NaN values with 0
kpi_analysis.fillna(0, inplace=True)
```

```python
kpi_analysis
```

```python
kpi_analysis.to_csv('../data/kpi_analysis_tableau.csv', index = False)
```

```python
kpi_analysis.to_excel('../data/kpi_analysis_tableau.xlsx', index = False)
```
