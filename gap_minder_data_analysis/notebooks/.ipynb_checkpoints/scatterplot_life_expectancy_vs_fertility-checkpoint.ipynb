```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 1
Open empty jupyter notebook and read in the dataframe that contains your gapminder data

```python
df = pd.read_csv('../data/gapminder_total.csv')
```

```python
df
```

### Step 2
Create subset of data for one year



```python
df_subset = df[df['year'] == 2000]
```

```python
df_subset
```

### Step 4
Plot the life expectancy vs fertility rate



```python
sns.scatterplot(data=df_subset,
            x='life expectancy', 
            y='fertility',
            hue='country',
            legend= False);
```

### Step 5
Using matplotlib add a title and labels. Increase the plot size to make it more readable.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df_subset,
            x='life expectancy', 
            y='fertility',
            hue='continent', 
            legend= 'brief');
plt.title('Life Expectancy vs Fertility Rate')
plt.ylabel('Fertility')
plt.xlabel('Life Expectancy');
```

### Step 6
Weight the size of the scatter points to the population of each country. This can be done using the size parameter in the sns.scatterplot() function.

Tip: the code above is correct but the magnitude of the population is astronomically larger than that of the life expectancy and fertility rate. Use some math to fix this.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df_subset,
            x='life expectancy', 
            y='fertility',
            hue='continent',
            size= df_subset['population']/100000,
            sizes=(50, 600)
               )
plt.legend(scatterpoints=1, frameon=False, labelspacing=0.8)
plt.title('Life Expectancy vs Fertility Rate')
plt.ylabel('Fertility')
plt.xlabel('Life Expectancy');
```

```python
def gapminder_plot(year):
    plt.figure(figsize=(8, 5), edgecolor='black')
    sns.set_style("darkgrid")
    
    # Filter the data for the current year
    df_year = df[df['year'] == year]
    
    # Creating scatter plot
    sns.scatterplot(data=df_year,
                    x='life expectancy',
                    y='fertility',
                    hue='continent',
                    size=df_year['population'] / 100000,
                    sizes=(50, 600),
                    alpha=0.8,
                    linewidths=200,  # Adjust the value to create more separation
                    )
    plt.xticks(rotation=45)
    plt.axis([0, 85, 0, 9.5])

    # Editing the legend
    handles, labels = plt.gca().get_legend_handles_labels()
    plt.legend(handles[0:7], labels[0:7], borderaxespad=0.7, fontsize=8, bbox_to_anchor=(1.4, 1))
    
    # Add labels and title
    plt.title('Life Expectancy vs Fertility Rate')
    plt.ylabel('Fertility')
    plt.xlabel('Life Expectancy')

    # Add label markers for specific countries
    countries_to_label = ['Pakistan', 'Germany', 'China', 'United States']
    for country in countries_to_label:
        x = float(df_year.loc[df_year['country'] == country, 'life expectancy'].values[0])
        y = float(df_year.loc[df_year['country'] == country, 'fertility'].values[0])
        plt.text(x, y, country, fontsize=8, ha='center', va='center')

    # Add year label
    plt.text(20, 9, f'Year: {year}', fontsize=8, ha='right', va='top', 
             bbox= dict(facecolor='whitesmoke', boxstyle='round,pad=0.5'))
    
   #editing the spine or the edge of the graph 
    for spine in plt.gca().spines.values():
        spine.set_visible(True)
        spine.set_linewidth(0.5)
        spine.set_color('gray')
    
    plt.tight_layout()
    plt.savefig(f'../scatterplots/plot_{year}.png', dpi=300)

```

```python
#call the function
gapminder_plot(2015)
```
