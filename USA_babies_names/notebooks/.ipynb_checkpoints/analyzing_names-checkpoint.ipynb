##### 4. How many entries are the in the 'Martin' DataFrame? How many years are in the DataFrame? Why are there more entries of the name 'Martin' than years? In how many years does the name 'Martin' appear for both genders?

```python
mask_martin = df['name'] ==  'Martin'
# df[mask_martin].shape
len(df[mask_martin])
```

```python
df_martin = df[mask_martin]
print(df_martin)
```

```python
unique_years = df_martin['year'].unique()
len(unique_years)
```

```python
martin_M_entries = df_martin[df_martin['gender']== 'M']
print(martin_M_entries)
```

```python
martin_F_entries = df_martin[df_martin['gender']== 'F']
print(martin_F_entries)
```

```python
both_gen = pd.merge(martin_F_entries,martin_M_entries, on='year')
print(both_gen)
```

#### # 5. Filter the combined DataFrame for a DataFrame which only has names that have a frequency over 2000

```python
freq_g_2000 = df['frequency'] > 2000
df[freq_g_2000]
```

#### 6. Filter the combined DataFrame for a DataFrame which only has names that have a frequency over 2000 and for years starting with 2000. What is the most common name since 2000?

```python
df_years_start_w_2000 = df['year'].between(2000,3000)
df[df_years_start_w_2000]
```

```python
comb_df_freq_year_2000 = df [ (freq_g_2000) & (df_years_start_w_2000 ) ]
comb_df_freq_year_2000
```

```python
m_common_name_f = comb_df_freq_year_2000['frequency'].max()
print(m_common_name_f)
```

```python
m_common_name = comb_df_freq_year_2000['frequency'] == m_common_name_f
comb_df_freq_year_2000[m_common_name]
```

#### 7. Filter the combined DataFrame for a DataFrame which only has names that have a frequency over 2000, for years starting with 2000 and for only females. What is the most common name given to females since 2000?

```python
only_F = comb_df_freq_year_2000['gender'] == 'F'
only_F_2000 = comb_df_freq_year_2000[only_F]
only_F_2000
```

```python
m_common_name_f_2000_freq = only_F_2000['frequency'].max()
print(m_common_name_f_2000)
```

```python
m_common_name_F_2000 = only_F_2000['frequency'] == m_common_name_f_2000_freq
only_F_2000[m_common_name_F_2000]
```
