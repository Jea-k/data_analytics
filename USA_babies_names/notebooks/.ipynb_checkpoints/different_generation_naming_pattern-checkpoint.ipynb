#### Identify naming patterns specific to different generations, such as
- Baby Boomers       (1946 - 1964)
- Generation X       (1965 - 1980)
- Millennials        (1981 - 1996)
- Generation Z       (1996 - 2020)

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../data/baby_names_all_years.csv')

baby_boomers_start = 1946
baby_boomers_end = 1964

gen_x_start = 1965
gen_x_end = 1980

millennials_start = 1981
millennials_end = 1996

gen_z_start = 1997
gen_z_end = 2020



baby_boomers = df[(df['year'] >= baby_boomers_start) & (df['year'] <= baby_boomers_end)]
gen_x = df[(df['year'] >= gen_x_start) & (df['year'] <= gen_x_end)]
millennials = df[(df['year'] >= millennials_start) & (df['year'] <= millennials_end)]
gen_z = df[(df['year'] >= gen_z_start) & (df['year'] <= gen_z_end)]

#Analyze the naming patterns within each generation

baby_boomers_common_names = baby_boomers['name'].value_counts().head(5)
gen_x_common_names = gen_x['name'].value_counts().head(5)
millennials_common_names = millennials['name'].value_counts().head(5)
gen_z_common_names = gen_z['name'].value_counts().head(5)

#plot used from the internet. need to figure it out later
fig, ax = plt.subplots(figsize=(15, 10))


ax.barh(baby_boomers_common_names.index, baby_boomers_common_names.values)
ax.barh(gen_x_common_names.index, gen_x_common_names.values)
ax.barh(millennials_common_names.index, millennials_common_names.values)
ax.barh(gen_z_common_names.index, gen_z_common_names.values)

legend_labels = ['Baby Boomers', 'Generation X', 'Millennials', 'Generation Z']
ax.legend(legend_labels)
ax.set_xlabel('Frequency')
ax.set_ylabel('Name')
ax.set_title('Most Common Names in Different Generations')

plt.show()
```

- Baby Boomers were influenced by traditional naming conventions
- Generation X has more diverse and unique names different from traditional norms
- Millennials have a mix of traditional names and modern names 
- Generation Z has a blend of traditional names, unique names, and names influenced by popular culture. Social media and internet trends have also influenced naming patterns for Generation Z.
