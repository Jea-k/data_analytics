#### Create a plot of the summed up frequency (total births) per year from 1880 to 2020.


```python
# births_dict = {'year':[], 'total_births':[]}
```

```python
#parse function
def parse_dataset(year):
    
    path = f'../data/yob{year}.txt'
    df = pd.read_csv(path, names = ['name','gender','frequency'])
#     df["year"] = year
    return df
```

```python
#Loop over the years from 1880 to 2020
births_dict = {'year':[], 'total_births':[]}
for year in range(1880, 2023):
    df = parse_dataset(year)
    sum_F = df['frequency'].sum()
    
    births_dict['total_births'].append(sum_F)
    births_dict['year'].append(year)

yearly_births = pd.DataFrame(births_dict)


  
```

###  Number of unique names over time

- How has the number of unique names given to newborn children varied over time?
- For each year from 1880 til 2020
     - read in the data using the parse_dataset() function
     - use nunique() to count the unique names for each year
     - keep track of the year and the count in a dictionary
     - convert dictionary to pandasDataFrame as was done in the first milestone for this lesson
- Visualize the data, use the years on the x-axis and the counts on the y-axis.

```python
unique_names = {'name':[]}

for year in range(1880, 2023):
    df_year = parse_dataset(year) 
    unique_names_count = len(df_year['name'].unique()) #used len() to solve ValueError: setting an array element with a sequence"
    unique_names['name'].append(unique_names_count) 


unique_names = pd.DataFrame(unique_names)
```

```python
plt.plot(range(1880,2023),
         unique_names['name'], 
         label = 'Number of unique names')
plt.title('Unique Names Given to Newborn Children over Time')
plt.ylabel('Count')
plt.xlabel('Year')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#function for distribution of name length over year

def name_length_dis(year):
    df_year = parse_dataset(year)
    df_year['name_length'] = df_year['name'].apply(lambda x: len(x))
    
    plt.hist(df_year['name_length'], 
          bins=20,
         label = year,
         edgecolor = 'k',
         alpha = 0.7,
         log=True)
    plt.title('Length of Names over year')
    plt.ylabel('Length')
    plt.legend()
    plt.show()

    return


    
```

```python
avg_name_lengths = {'year': [], 'avg_name_length': []}

for year in range(1880, 2023):
    df_year = parse_dataset(year)
    name_length = df_year['name'].apply(lambda x: len(x))
    avg_name_length = name_length.mean()
    avg_name_lengths['year'].append(year)
    avg_name_lengths['avg_name_length'].append(avg_name_length)


df_avg_name_length = pd.DataFrame(avg_name_lengths)

df_avg_name_length
```

```python
plt.plot(df_avg_name_length['year'], df_avg_name_length['avg_name_length'], label = 'Average name length')
plt.title('Average Length of Names Over Time')
plt.xlabel('Year')
plt.ylabel('Length')
plt.grid(True)
plt.legend()
plt.show()
```
