Battery Temperature Improvement:
Create a scatter plot with 'Battery Temperature (Start)' on the x-axis and 'Battery Temperature (End)' on the y-axis. Each point represents a trip, and if battery technology has advanced, you should see a trend where th


```python
import plotly.express as px
# Create a scatter plot
fig = px.line(df_bat, x='Battery Temperature (Start) [°C]', y='Battery Temperature (End)',
                 title='Battery Temperature Improvement',
                 labels={'Battery Temperature (Start) [°C]': 'Battery Temperature (Start) [°C]',
                         'Battery Temperature (End)': 'Battery Temperature (End) [°C]'})

# Add a trendline for visual reference
fig.update_traces(mode='markers+lines', marker=dict(size=8))

# Show the plot
fig.show()
```
