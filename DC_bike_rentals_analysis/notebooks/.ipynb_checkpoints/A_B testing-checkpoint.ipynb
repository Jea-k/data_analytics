### A/B Testing

Bike dataset has the information about the weather that is stored in columns such as weather, temp, atemp, humidity, windspeed. Create a notebook where you perfom a hypothesis testing that checks if there is a significant difference between two of the temperature columns: temp and atemp. Use the scipy.stats and t-test statistics.

```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('../data/bike_dataset_new.csv',parse_dates=['datetime'])
df
```

```python
temp = df["temp"]
atemp = df["atemp"]
```

```python
test_statistic, pvalue = stats.ttest_ind(temp, atemp)
print ('test_statistic = ', test_statistic)
print ('pvalue = ', pvalue)
```

```python
alpha = 0.05 
#significance value: 5% of the times we will be able to reject the null hypothesis and 95% we will not
```

```python
if pvalue < alpha:
    print('The H0 is rejected and there is a difference between temp and atemp')
else:
    print('Failed to reject H0 and there is no difference between temp and atemp')
```
