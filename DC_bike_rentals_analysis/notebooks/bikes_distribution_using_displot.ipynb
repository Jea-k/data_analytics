##  Bikes Distribution

Create a notebook that is solving following questions:

1. Filter bike data for one day
2. Using sns.displot() plot the distribution of the count during that day (Try setting the parameter kde=True). More info on this plotting function here: displot
3. Filter bike data for the year 2011
4. Again using sns.displot() plot the distribution of the temperature of the entire year

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import statsmodels.formula.api as sm
```

```python
df = pd.read_csv('../data/bike_dataset_new.csv',parse_dates=['datetime'])
df
```

#### 1. Filter bike data for one day

```python
df_day= df [ df['day'] == 7]
df_day
```

```python
df_day['count'].max()
```

#### 2.Using sns.displot() plot the distribution of the count during that day (Try setting the parameter kde=True). More info on this plotting function here: displot

```python
sns.set(style='dark')
sns.displot(data = df_day, 
            x= 'count',
            kde=True, #to enable the Kernel Density Plot along with the distplot
            stat="probability")
plt.title('Distribution of Traffic Density for 1 day');
```

- x-axis is distribution of actual data from the dataset
- y-axis is the density distribution based on probability / likelihood of differenct values occuring

#### 3. Filter bike data for the year 2011

```python
df_year = df [ df['year'] == 2011]
df_year
```

#### 4. Again using sns.displot() plot the distribution of the temperature of the entire year

```python
sns.set(style='dark')
sns.displot(data = df_year, 
            x= 'temp',
            kde=True,
           stat="probability")
plt.title('Distribution of Temperature for 2011');
```
